```python
import matplotlib.pyplot as plt
x=["python","c++","c","java"]
y=[80,70,50,90]
plt.xlabel("languages",fontsize=10)
plt.ylabel("no",fontsize=10)
plt.title("Barchart",fontsize=15)
plt.bar(x,y)
plt.show
```

```python
import matplotlib.pyplot as plt

x = ["python", "c++", "c", "java"]
y = [80, 70, 50, 90]
z=[20,30,40,50]
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Number", fontsize=10)
plt.title("Bar Chart", fontsize=15)
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,z,width=0.4,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = ["python", "c++", "c", "java"]
y = [80, 70, 50, 90]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.2
p1=[j+width for j in p]
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Number", fontsize=10)
plt.title("Bar Chart", fontsize=15)
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.legend()
plt.show()
```
