```python
import matplotlib.pyplot as plt

x = ["python", "c++", "c", "java"]
y = [80, 70, 50, 90]
plt.xlabel("Languages", fontsize=10)
plt.ylabel("Number", fontsize=10)
plt.title("Bar Chart", fontsize=15)
colors = ["y", "g", "b", "m"]
plt.bar(x,y,width=0.4,color=colors,align="edge",edgecolor="black",linewidth=5,linestyle="dotted")
plt.show()
```
